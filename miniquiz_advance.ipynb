```python
import matplotlib.pyplot as plt

# Data untuk plot pertama
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 1, 3, 5]

# Data untuk plot kedua
x2 = [1, 2, 3, 4, 5]
y2 = [5, 3, 1, 4, 2]

# Membuat tata letak subplot
fig, axs = plt.subplots(2, 2)

# Menggambar plot pertama pada subplot di atas kiri
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Plot Pertama')

# Menggambar plot kedua pada subplot di atas kanan
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Plot Kedua')

# Menggambar plot ketiga pada subplot di bawah kiri
axs[1, 0].plot(y1, x1)
axs[1, 0].set_title('Plot Ketiga')

# Menggambar plot keempat pada subplot di bawah kanan
axs[1, 1].plot(y2, x2)
axs[1, 1].set_title('Plot Keempat')

# Menampilkan plot
plt.show()
```

Dalam boxplot, distribusi data dapat dikatakan **uniform** *jika lebar kotak (box) hampir sama besar dan median (garis tengah kotak) berada di tengah-tengah kotak*. Selain itu, jarak antara kuartil bawah (Q1) dan kuartil atas (Q3) juga harus hampir sama besar.

 Jika boxplot memiliki bentuk yang mirip dengan huruf "U", maka distribusinya dapat disebut U-shaped.
 Jika distribusi data memiliki bentuk U-shaped, maka boxplot akan menunjukkan ciri-ciri sebagai berikut:

1. Kotak akan berada pada posisi yang lebih rendah, menunjukkan sebagian besar data berada di nilai-nilai rendah.
2. Median akan terletak di dekat tengah kotak.
3. Whisker yang menghubungkan kotak ke titik-titik di atas dan di bawah akan lebih panjang daripada pada distribusi yang simetris.
4. Ada kemungkinan adanya outlier pada kedua ujung whisker, menunjukkan adanya data yang sangat rendah atau sangat tinggi.

```python
import matplotlib.pyplot as plt
import numpy as np

# generate data
data = np.concatenate([np.random.normal(10, 2, 100), np.random.normal(20, 2, 100), np.random.normal(10, 2, 100)])

# create boxplot
fig, ax = plt.subplots(figsize=(6,8))
ax.boxplot(data, vert=False, widths=0.7)

# add title and labels
ax.set_title('Contoh Boxplot U-shaped')
ax.set_xlabel('Nilai Data')

# show plot
plt.show()
```

3. Setelah itu, kita membuat boxplot menggunakan fungsi boxplot() dengan argumen vert=False untuk menampilkan boxplot secara horizontal, dan widths=0.7 untuk mengatur lebar kotak.
4. Terakhir, kita menambahkan judul dan label sumbu x menggunakan fungsi set_title() dan set_xlabel(), dan menampilkan plot menggunakan fungsi show().

Distribusi dari boxplot dapat disebut **skewed (miring)** jika median tidak berada di tengah kotak (box) dan bagian atas kotak lebih pendek dari bagian bawahnya, atau sebaliknya. Hal ini menunjukkan bahwa data tidak memiliki distribusi yang simetris.

```python
import matplotlib.pyplot as plt
import numpy as np

# generate data
data = np.concatenate([np.random.normal(5, 1, 50), np.random.normal(10, 2, 50), np.random.normal(15, 3, 50)])

# create boxplot
fig, ax = plt.subplots(figsize=(6,8))
ax.boxplot(data, vert=False, widths=0.7)

# add title and labels
ax.set_title('Contoh Boxplot Skewed')
ax.set_xlabel('Nilai Data')

# show plot
plt.show()
```

* Pertama-tama, kita mengimport library Matplotlib dan NumPy.
* Kemudian, kita membuat data yang terdiri dari tiga kelompok data dengan rata-rata yang berbeda-beda.
* Setelah itu, kita membuat boxplot menggunakan fungsi boxplot() dengan argumen vert=False untuk menampilkan boxplot secara horizontal, dan widths=0.7 untuk mengatur lebar kotak.
* Terakhir, kita menambahkan judul dan label sumbu x menggunakan fungsi set_title() dan set_xlabel(), dan menampilkan plot menggunakan fungsi show().

Distribusi dari boxplot dapat disebut **normal** jika median berada di tengah kotak (box) dan whisker atas dan bawah memiliki panjang yang seimbang atau hampir sama. Hal ini menunjukkan bahwa data memiliki distribusi yang simetris dan seragam.

```python
import matplotlib.pyplot as plt
import numpy as np

# generate data
data = np.random.normal(0, 1, 100)

# create boxplot
fig, ax = plt.subplots(figsize=(6,8))
ax.boxplot(data, vert=False)

# add title and labels
ax.set_title('Contoh Boxplot Distribusi Normal')
ax.set_xlabel('Nilai Data')

# show plot
plt.show()
```

Distribusi dari boxplot dapat disebut **uniform** jika lebar kotak (box) hampir sama dengan lebar whisker atas dan bawah, dan kedua whisker memiliki panjang yang sama. Hal ini menunjukkan bahwa data memiliki distribusi yang seragam dan *tidak memiliki outlier.*

```python
import matplotlib.pyplot as plt
import numpy as np

# generate data
data = np.random.uniform(0, 10, 100)

# create boxplot
fig, ax = plt.subplots(figsize=(6,8))
ax.boxplot(data, vert=False)

# add title and labels
ax.set_title('Contoh Boxplot Distribusi Uniform')
ax.set_xlabel('Nilai Data')

# show plot
plt.show()
```

```python
import pandas as pd

# buat data frame
df = pd.DataFrame({
    'Age': [10, 15, 18, 22, 25, 28, 32, 35, 40, 45],
    'Income': [1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500]
})

# buat interval batas umur
bins_age = [0, 18, 25, 35, 50]

# buat interval batas penghasilan
bins_income = [0, 2000, 4000, 6000]

# kelompokkan data umur ke dalam interval
df['AgeGroup'] = pd.cut(df['Age'], bins_age, labels=['<18', '18-25', '25-35', '35+'])

# kelompokkan data penghasilan ke dalam interval
df['IncomeGroup'] = pd.cut(df['Income'], bins_income, labels=['<2000', '2000-4000', '4000+'])

# tampilkan data frame
print(df)
```
